```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# Importing CSV files - different seeds
impactFactorMarked = pd.read_csv('impactFactorMarked.csv')
impactFactorMarked = impactFactorMarked.drop(['Unnamed: 0'], axis = 1)

impactFactorGCR = pd.read_csv('impactFactorGCR.csv')
impactFactorGCR = impactFactorGCR.drop(['Unnamed: 0'], axis = 1)
```

```python
style.use('fivethirtyeight')

graph = impactFactorMarked.boxplot(grid = False, figsize = (12, 8))

# Setting rules for labeling axes
graph.tick_params(axis = 'both', which = 'major', labelsize = 12)

# Title

# Y value is continuous to values on Y-axis. For instance, 
# if the maximum value is 60, the text can begin at 61.

# Alpha indicates text opacity

graph.text(x = 1, y = 1.2, s = "Impact Factor Results - Marked Images",
               fontsize = 24, weight = 'bold', alpha = .7)

# Subtitle
graph.text(x = 1, y = 1.1,
               s = "...\n...",
              fontsize = 14, alpha = .6)
```

```python
style.use('fivethirtyeight')

graph = impactFactorGCR.boxplot(grid = False, figsize = (12, 8))

# Setting rules for labeling axes
graph.tick_params(axis = 'both', which = 'major', labelsize = 12)

# Title

# Y value is continuous to values on Y-axis. For instance, 
# if the maximum value is 60, the text can begin at 61.

# Alpha indicates text opacity

graph.text(x = 1, y = 1.2, s = "Impact Factor Results - GCR Images",
               fontsize = 24, weight = 'bold', alpha = .7)

# Subtitle
graph.text(x = 1, y = 1.1,
               s = "...\n...",
              fontsize = 14, alpha = .6)
```

```python
# Colors: https://github.com/matplotlib/matplotlib/blob/38be7aeaaac3691560aeadafe46722dda427ef47/lib/matplotlib/mpl-data/stylelib/fivethirtyeight.mplstyle
plt.style.use('fivethirtyeight')

# Creating graph
fig, ax1 = plt.subplots(figsize = (12,8))

# Setting X-axis and first Y-axis
ax1.set_xlabel('Impact Factor', color = '#000000', fontsize = 14)
ax1.set_ylabel('Correlation', color = '#000000', fontsize = 14)

# Creating a second Y-axis that shares the same X-axis
ax2 = ax1.twinx()
ax2.set_ylabel('', color = '#000000', fontsize = 14)

# Creating a second X-axis that shares the same Y-axis
ax2 = ax1.twiny() 
ax2.set_xlabel('', color = '#000000', fontsize = 14) 

# Show grid
ax1.grid(False, which='major', axis='both')
ax2.grid(True, which='major', axis='both')

# Setting different colors for each variable
xRange = range(0, 1501, 100)

ax1.plot(xRange, medianAll['Marked'], color = '#888888')
ax2.plot(xRange, medianAll['GCR'], color = '#c1c1c1')
```

```python
plt.style.use('fivethirtyeight')
    
medianAll.plot(x = "Impact Factor", y = ["Marked", "GCR"], figsize = (12, 8))

# Setting top and bottom values for Y-axes
plt.xaxis.set_visible(False)
plt.ylim(bottom=0, top=1)

# Saving figure as EPS
PlotName = 'plots/ImpactFactor.eps'
plt.savefig(PlotName, format="eps")

# Displaying plot
plt.show()
```
